```python
import pandas as pd
from sqlalchemy import create_engine
from config import db_password
```

```python
abortions_df = pd.read_excel("GuttmacherDataCenter.xlsx", 
                                       skiprows=[52,53,54], 
                                       index_col='U.S. State')
abortions_df = abortions_df.rename(columns={"U.S. State": "State",
                                            "% of all U.S. abortions, by state of occurrence, 2020":"Percent_of_all_US_abortions", 
                                           "% of residents obtaining abortions who traveled out of state for care, 2020": "Percent_residents_traveling_outside_state",
                                           "No. of abortions, by state of occurrence, 2020 [1]": "abortions_occurring_state",
                                           "No. of abortions, by state of residence, 2020 [2]": "abortions_residence_state"})
abortions_df
```

```python
abortions_df['abortions_occurring_state'] = abortions_df['abortions_occurring_state'].str.replace(',', '')
```

```python
abortions_df['abortions_residence_state'] = abortions_df['abortions_residence_state'].str.replace(',', '')
```

```python
abortions_df.index.names = ['State']
```

```python
abortions_df
```

```python
pd.DataFrame.to_csv(abortions_df, 'abortions.csv', sep=',', na_rep='.', index=True)
```

```python
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5433/Reproducitve_Access"
engine = create_engine(db_string)
abortions_df.to_sql(name='abortions_data', con=engine)
```
